```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import zipfile
import os
```

```python
# For display settings
pd.set_option('display.max_columns', None)
```

```python
# location of the ZIP file
file_path = r"C:\Users\tanzi\Personal Projects\Marketing-Campaign-Performance-Analysis\Marketing-Campaign-Performance-Analysis\data\cleaned_campaign_data.zip"

# Opening and laoding into dataframe
with zipfile.ZipFile(file_path, 'r') as z:
    with z.open('cleaned_campaign_data.csv') as f:
        df = pd.read_csv(f)
```

```python
df.head()
```

```python
# Display summary statistics
df.describe(include="all")
```

```python
# Average ROI & Conversion Rate by Marketing Channel
campaign_performance = df.groupby("Channel_Used")[["ROI", "Conversion_Rate"]].mean().sort_values(by="ROI", ascending=False)
```

```python
# Plot
plt.figure(figsize=(10, 5))
sns.barplot(x=campaign_performance.index, y=campaign_performance["ROI"], palette="viridis")
plt.title("Average ROI by Marketing Channel", fontsize=14)
plt.ylabel("ROI")
plt.xlabel("Marketing Channel")
plt.xticks(rotation=45)
plt.show()
```

```python
display(campaign_performance[["ROI"]])
```

```python
numeric_columns = ['Conversion_Rate', 'Acquisition_Cost', 'ROI', 'Clicks', 'Impressions', 'Engagement_Score', 'Duration_Days', 'CPC', 'CTR']
df_numeric = df[numeric_columns]

# Heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(df_numeric.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x="Customer_Segment", order=df["Customer_Segment"].value_counts().index, palette="magma")
plt.xticks(rotation=45)
plt.title("Campaign Distribution by Customer Segment")
plt.show()
```

```python
display(df["Customer_Segment"].value_counts().reset_index().rename(columns={"index": "Customer_Segment", "Customer_Segment": "Count"}))
```

```python
df["Year-Month"] = df["Date"].str[:7]  # Extract Year-Month
roi_trend = df.groupby("Year-Month")["ROI"].mean()

plt.figure(figsize=(12,5))
sns.lineplot(data=roi_trend, marker="o")
plt.xticks(rotation=45)
plt.title("ROI Trends Over Time")
plt.xlabel("Time (Year-Month)")
plt.ylabel("Average ROI")
plt.show()
```

```python
# Define paths
cleaned_file_path = r"C:\Users\tanzi\Personal Projects\Marketing-Campaign-Performance-Analysis\Marketing-Campaign-Performance-Analysis\data\campaign_summary.csv"
zip_file_path = r"C:\Users\tanzi\Personal Projects\Marketing-Campaign-Performance-Analysis\Marketing-Campaign-Performance-Analysis\data\campaign_summary.zip"

# Save the cleaned dataset
df.to_csv(cleaned_file_path, index=False)

# Zip the cleaned dataset with the correct CSV name
with zipfile.ZipFile(zip_file_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
    zipf.write(cleaned_file_path, arcname="campaign_summary.csv")  # ✅ Corrected

# Remove the original cleaned CSV file (optional, to save space)
os.remove(cleaned_file_path)
```
